```python
def needleman(seq1, seq2):
    match = 2; mismatch = -1; gap = -2; m, n = len(seq1),  len(seq2)
    dp = []
    for i in range(m + 1):
        row = []
        for j in range(n + 1):
            row.append(0)
        dp.append(row)
    for i in range(m + 1):
        dp[i][0] = i * gap
    for j in range(n + 1):
        dp[0][j] = j * gap
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if seq1[i - 1] == seq2[j - 1]:
                score = match
            else:
                score = mismatch
            diag = dp[i - 1][j - 1] + score
            up = dp[i - 1][j] + gap
            left = dp[i][j - 1] + gap
            dp[i][j] = max(diag, up, left)
    return dp[m][n]
print("The score for Needleman-Wunsch is: ", needleman("GAATTC", "GATAC"))
```

```python
def smith(seq1, seq2):
    match = 2; mismatch = -1; gap = -2; m, n = len(seq1), len(seq2)
    dp = []
    for i in range(m + 1):
        row = []
        for j in range(n + 1):
            row.append(0)
        dp.append(row)
    best = 0
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if seq1[i - 1] == seq2[j - 1]:
                score = match
            else:
                score = mismatch
            diag = dp[i - 1][j - 1] + score
            up = dp[i - 1][j] + gap
            left = dp[i][j - 1] + gap
            value = max(0, diag, up, left)
            dp[i][j] = value
            if value > best:
                best = value
    return best
print("The score for Smith-Waterman is: ", smith("GAATTC", "GATAC"))
```
